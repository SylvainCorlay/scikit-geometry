```python
import pygal
from pygal import draw

from matplotlib import pyplot as plt
import matplotlib.patches as patches

import math
```

```python
a = pygal.Point2(4, 5)
b = pygal.Point2(2, 1)
c = pygal.Point2(3, 6)
d = pygal.Point2(5,2)
print('a:', a)
```

### Point_2

```python
draw.draw_point(a, color='purple')
```

### Point_list

```python
point_list = [a,b,c,d]
print(point_list)
```

```python
draw.draw_list(point_list, color = 'red')
```

### Vector_2

```python
vector = pygal.Vector2(a, c)
print(vector)
```

```python
draw.draw_vector(vector, start=-1.5, stop=1.5, color='blue')
```

### Ray_2

```python
ray = pygal.Ray2(a, d)
print(ray)
```

```python
draw.draw_ray(ray, stop=10, color='red')
```

### Direction_2

```python
direction = pygal.Direction2(vector)
print(direction)
```

```python
draw.draw_direction(direction, start=-2, stop=2, color='orange')
```

### Bbox_2

```python
bbox = a.bbox() + b.bbox() + c.bbox()
print(bbox)
```

```python
draw.draw_bbox(bbox, display_range=3, fill=True, color='green')
```

### Line_2

```python
line = pygal.Line2(a, b)
print(line)
```

```python
draw.draw_line(line, start=-6, stop=6, num=150, color='black')
```

### Iso_rectangle_2

```python
iso_rectangle = pygal.IsoRectangle2(a,b,c,d)
print(iso_rectangle)
```

```python
draw.draw_iso_rectangle(iso_rectangle, display_range=5, fill=True, color='grey')
```

### Segment_2

```python
segment = pygal.Segment2(a,b)
print(segment)
```

```python
draw.draw_segment(segment, color='green')
```

### Circle

```python
circle = pygal.Circle2(a,b,c)
print(circle)
```

```python
draw.draw_circle(circle, display_range=4, fill=True, color='blue')
```

### Polygon

```python
polygon = pygal.Polygon([a, c, b, d])
print(polygon)
```

```python
draw.draw_polygon(polygon, facecolor='blue', point_color='red')
```

```python
e = pygal.Point2(3,2)
f = pygal.Point2(4,2)
g = pygal.Point2(4,4)
h = pygal.Point2(3,4)
```

### PolygonWithHoles

```python
polygon2 = pygal.Polygon([e, f, g, h])
polygon_with_holes = pygal.PolygonWithHoles(polygon, [polygon2])
print(polygon_with_holes)
```

```python
draw.draw_polygon(polygon, polygon_with_holes, facecolor='blue', point_color='red')
```
