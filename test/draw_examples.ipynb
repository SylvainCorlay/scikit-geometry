```python
import pygal
from pygal.draw import draw
from pygal import voronoi

from matplotlib import pyplot as plt
import matplotlib.patches as patches

import math
```

```python
a = pygal.Point2(4, 5)
b = pygal.Point2(2, 1)
c = pygal.Point2(3, 6)
d = pygal.Point2(5,2)
print('a:', a)
```

### Point_2

```python
draw(a, color='purple')
```

### Point_list

```python
point_list = [a,b,c,d]
print(point_list)
```

```python
draw(point_list, color = 'red')
```

### Vector_2

```python
vector = pygal.Vector2(a, c)
print(vector)
```

```python
draw(vector, start=-1.5, stop=1.5, color='blue')
```

```python
draw(vector, start=-1.5, stop=1.5, color='blue')
draw(vector.perpendicular(pygal.Sign.COUNTERCLOCKWISE), start=-1.5, stop=1.5, color='orange')
```

### Ray_2

```python
ray = pygal.Ray2(a, d)
print(ray)
```

```python
draw(ray, stop=10, color='red')
```

### Direction_2

```python
direction = pygal.Direction2(vector)
print(direction)
```

```python
draw(direction, start=-2, stop=2, color='orange')
```

### Bbox_2

```python
bbox = a.bbox() + b.bbox() + c.bbox()
print(bbox)
```

```python
draw(bbox, display_range=3, fill=True, color='green')
```

```python
bbox = a.bbox() + b.bbox() + c.bbox() + d.bbox()

draw(bbox, display_range=3, fill=True, color='green', alpha=0.2)
draw([a,b,c,d])
```

### Line_2

```python
line = pygal.Line2(a, b)
print(line)
```

```python
draw(line, start=-6, stop=6, num=150, color='black')
```

### Iso_rectangle_2

```python
iso_rectangle = pygal.IsoRectangle2(b, a, d, c)
print(iso_rectangle)
```

```python
draw(iso_rectangle, display_range=5, fill=True, color='grey')
```

### Segment_2

```python
segment = pygal.Segment2(c,d)
print(segment)
```

```python
draw(segment, color='green')
```

```python
pinter = pygal.intersection(line, segment)
draw(segment)
draw(line)
draw(pinter, color='red')
```

```python
segment = pygal.Segment2(a, b)
pinter = pygal.intersection(line, segment)
draw(segment)
draw(line)
draw(pinter, color='red')
```

### Circle

```python
circle = pygal.Circle2(a,b,c)
print(circle)
```

```python
draw(circle, display_range=4, fill=True, color='blue')
```

### Polygon

```python
polygon = pygal.Polygon([a, c, b, d])
print(polygon)
```

```python
draw(polygon, facecolor='lightblue', point_color='red')
```

```python
e = pygal.Point2(3,2)
f = pygal.Point2(4,2)
g = pygal.Point2(4,4)
h = pygal.Point2(3,4)
```

### PolygonWithHoles

```python
polygon2 = pygal.Polygon([e, f, g, h])
polygon_with_holes = pygal.PolygonWithHoles(polygon, [polygon2])
print(polygon_with_holes)
```

```python
draw(polygon, polygon_with_holes, facecolor='lightblue', point_color='red')
```

### Voronoi Diagram

```python
point_list+= [e, f, g, h]
voronoi_diagram = voronoi.VoronoiDiagram()
for point in point_list:
    voronoi_diagram.insert(point)
print(voronoi_diagram)
```

```python
draw(voronoi_diagram, sites_color='blue', finite_edges_color='red')
```
